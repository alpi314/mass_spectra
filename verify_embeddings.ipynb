## This Juptyer notebook is used to verify the embeddings generated by the model

```python
import pandas as pd
```

CSV File will include title and inchikey columns. Other columns are enumerated from 0 to N. Each column represent a feature in the embedding space.  
To check the embedding consistency, we will check deviations inside the same class and between different classes. Classes will be defined by the inchikey.

```python
FILE = "./embeddings/tms_new_spec2vec_embedding.csv"
df = pd.read_csv(FILE)
df = df.drop(columns=["title"])
```

```python
df.head()
```

```python
# group byinchi key and take the mean of the embeddings
inchi_mean = df.groupby("inchikey").mean()
inchi_mean.head()
```

```python
# get inclass standard deviation
inchi_std = df.groupby("inchikey").std()
inchi_std.head()
```

```python
# get between class standard deviation
inchi_std = inchi_mean.std()
inchi_std
```

```python
from matchms import calculate_scores
from spec2vec import Spec2Vec
from spec2vec_train import preprocess_file
import gensim

MODEL_FILE = "./models/tms/spec2vec.model"
model = gensim.models.Word2Vec.load(MODEL_FILE)
spec2vec_similarity = Spec2Vec(model=model, intensity_weighting_power=0.5,
                               allowed_missing_percentage=5.0)
```

```python
spectrum_documents = preprocess_file("./dataset/Test dataset_TMS_RAW.mgf")
```

```python
scores = calculate_scores(spectrum_documents, spectrum_documents, spec2vec_similarity, is_symmetric=True)
```

```python
counts_by_inchikey = {}
for s in spectrum_documents:
    inchikey = s.metadata.get("inchikey")
    if inchikey not in counts_by_inchikey:
        counts_by_inchikey[inchikey] = 0
    counts_by_inchikey[inchikey] += 1
```

```python
# create confusion matrix
confusion_matrix = pd.DataFrame(index=counts_by_inchikey.keys(), columns=counts_by_inchikey.keys())

# initialize all values to 0
for col in confusion_matrix.columns:
    confusion_matrix[col].values[:] = 0

for i, s in enumerate(spectrum_documents):
    target_inchi = s.metadata.get("inchikey") # expected inchikey
    expected_equal_inchikeys = counts_by_inchikey[target_inchi] # expected number of equal inchikeys
    best_matches = scores.scores_by_query(s, sort=True)[:expected_equal_inchikeys] # get the best matches

    # we expect all matches to have the same inchikey
    for j, match in enumerate(best_matches):
        inchikey = match[0].metadata.get("inchikey") # actual inchikey of the match
        confusion_matrix.loc[target_inchi, inchikey] += 1 # add 1 to the confusion matrix
```
